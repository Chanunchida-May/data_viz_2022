```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv') #ข้อมูลของดอกไม้ไอริส
df
```

```python
df['Name'].unique() #เรียกดูชื่อพันธุ์ดอกไม้
```

#Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df,'Name')
```

```python
pd.plotting.parallel_coordinates(df,'Name',cols = ['SepalLength','PetalLength']) #cols เป็นตัวเเปรที่ระบุว่าเราจะใช้คอลัมน์อะไรมาพอตบ้าง
```

```python
pd.plotting.parallel_coordinates(df,'Name',cols = ['SepalLength','PetalLength']); 
```

##Change color

```python
pd.plotting.parallel_coordinates(df,'Name',color=['r','b','g'])  #เปลี่ยนสีเส้นในกราฟ
```

```python
pd.plotting.parallel_coordinates(df,'Name',color=['darkolivegreen','teal','darkorchid']) #เปลี่ยนสีเส้นในกราฟ
```

```python
pd.plotting.parallel_coordinates(df,'Name',color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]]) #ปรับตามค่าสีปริมาณเเสงที่กำหนดไว้ว่ามากน้อยเเค่ไหน
```

#Specify color by Colormap
https://matplotlib.org/stable/tutorials/colors/colormaps.html



```python
pd.plotting.parallel_coordinates(df,'Name',colormap='Set2')
```

##Heatmap matrix

```python
matplotlib.rcParams['figure.figsize']=[25,25] #กำหนดขนาดของfigure
plt.imshow(df.iloc[:,:-1],interpolation='nearest') 
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap = 'copper') #สีเข้ม = ค่าน้อย , สีสว่าง = ค่ามาก
```

```python
plt.subplot(1,3,1) #subplot(มีกี่เเถว,มีกี่หลัก,เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='afmhot') 
plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='afmhot') 
plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='afmhot') 
```

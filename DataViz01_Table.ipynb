```python
data_covid = pd.read_csv(data)  # pd.read_csv เป็นการอ่านข้อมูล แล้วตั้งชื่อตัวแปรที่เก็บไว้ว่า data_covid
data_covid
```

จากการตรวจสอบพบว่ายังมีค่า missing อยู่

```python
data_con = data_vid.dropna() # สร้างตัวแปร data_con โดยตัดค่า missing ออกจาก data_vid
data_con # แสดงผล
```

```python
data_con.isnull().any() # ตรวจสอบค่าmissing ในตาราง data_con อีกครั้ง
```

จากการตรวจสอบพบว่าไม่มีค่า missing 

```python
set(data_con['risk']) # ใช้คำสั่ง set เพื่อต้องการดูรายการทั้งหมดในตาราง data_con คอลัมน์ risk
```

#risks ที่เลือก Cluster ระยอง พิธีกรรมทางศาสนาและไปสถานที่ชุมชน เช่น ตลาดนัด สถานที่ท่องเที่ยว

```python
data_HW1 = data_con[data_con['risk']=='Cluster ระยอง'] #สร้างตัวแปร data_HW1 เพื่อดึงข้อมูลในตาราง data_con เลือกคอลัมน์ risk เฉพาะ Cluster ระยองมา
data_HW1 # แสดงผล
```

```python
data_HW2 = data_con[data_con['risk']=='พิธีกรรมทางศาสนา'] #สร้างตัวแปร data_HW2 เพื่อดึงข้อมูลในตาราง data_con เลือกคอลัมน์ risk เฉพาะ พิธีกรรมทางศาสนามา
data_HW2 # แสดงผล
```

```python
data_HW3 = data_con[data_con['risk']=='ไปสถานที่ชุมชน เช่น ตลาดนัด สถานที่ท่องเที่ยว'] #สร้างตัวแปร data_HW3 เพื่อดึงข้อมูลในตาราง data_con เลือกคอลัมน์ risk เฉพาะ ไปสถานที่ชุมชน เช่น ตลาดนัด สถานที่ท่องเที่ยวไปสถานที่ชุมชน มา
data_HW3 # แสดงผล
```

```python
data_HW =pd.concat([data_HW1,data_HW2,data_HW3]) #สร้างตัวแปร data_HW ทำการ concat โดยต่อตารางแกน X ตาราง data_HW1,data_HW2,data_HW3
data_HW # แสดงข้อมูล
```

```python
data_Man = data_HW.replace('ชาย',10) # สร้างตัวแปร data_Man ใช้คำสั่ง replace โดยให้ 10 แทนเพศชาย ในคอลัมน์ sex 
data_Man #แสดงผล
```

```python
data_final1 = data_Man.replace('หญิง',90) # สร้างตัวแปร data_final1 ใช้คำสั่ง replace โดยให้ 90 แทนเพศหญิง ในคอลัมน์ sex เพื่อเปลี่ยนเพศหญิงที่เหลือจากเพศชายในตาราง data_Man
data_final1 #แสดงผล
```

```python
data_final1['risk'].unique() #ตาราง data_final1 ต้องการดูเฉพาะคอลัมน์ risk ที่เราเลือก ที่มี3 risk มีอะไรบ้าง โดยเลือกค่ามาทีละ 1
```

```python
pd.plotting.parallel_coordinates(data_final1, 'risk',colormap ='Set2') ; #ใช้คำสั่งพล็อตกราฟ parallel coordinate  ตาราง data_final1 ที่มี 3 risk และเลือกสีเป็น Set2
```

```python
data_HW_G = data_final1.replace({'Cluster ระยอง': 'Cluster Rayong','พิธีกรรมทางศาสนา': 'Religious ritual' ,'ไปสถานที่ชุมชน เช่น ตลาดนัด สถานที่ท่องเที่ยว':'go to community place'}) # สร้างตัวแปร data_HW_G ใช้คำสั่ง replace โดยแทนชื่อทั้ง3 risk เป็นภาษาอังกฤษแทนชื่อไทย เพื่อสร้างกราฟชื่อจะปรากฏในกราฟ 
data_HW_G #แสดงผล
```

```python
data_HW_G['risk'].unique() # ตาราง data_HW_G ต้องการดูเฉพาะคอลัมน์ risk ที่เราเลือก ที่มี3 risk มีอะไรบ้าง โดยเลือกค่ามาทีละ 1
```

```python
pd.plotting.parallel_coordinates(data_HW_G, 'risk',colormap = 'Set2' ) ; #ใช้คำสั่งพล็อตกราฟ parallel coordinate  ตาราง data_HW_G ที่มี 3 risk และเลือกสีเป็น Set2
```

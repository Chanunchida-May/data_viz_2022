###HW กลุ่ม 2 เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

````
a_date =list() # สร้าง list เปล่าเพื่อนำไปเก็บไว้ที่ตัวแปร a_data 
for each_row in data_covid.iterrows(): # ไปดูทุกแถวในตาราง data_covid 
    if each_row[1]['announce_date'].split('/')[1] <= '20': #เป็นการดูแต่ละแถว, [1]['announce_date'] คือการเรียกดูข้อมูลเริ่มจากตัวที่ 1 ของแต่ละแถว และดูที่คอลัมน์ announce_date และแบ่งค่าข้อมูลโดยตัดตามเครื่องหมาย / เมื่อตัดแล้ว ตัวที่1 ของคอลัมน์ announce_date ต้องมีค่าน้อยกว่าหรือเท่ากับ 20
        a_date.append(True) # ถ้าข้อมูลตรงตามเงื่อนไข  จะทำการเพิ่มข้อมูลเข้าไปใน list a_date
    else:
        a_date.append(False) # ถ้าข้อมูลไม่ตรงตามเงื่อนไข จะไม่เพิ่มข้อมูลเข้าไปใน list

data_covid[a_date] # แสดงค่าข้อมูลตาราง data_covid 
````
